```python
Region = df[["Region"]].groupby('Region').sum()
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
reg = df['Region'].value_counts().index[:5]
reg1 = df['Region'].value_counts().index[:5]
dic ={}
for each in range(0,5):
  dic[reg[each]] = reg1[each]
```

```python
def plot(area,name):
    middleEastTerrorism = df[df['Region'] == area]
    middleEastTerrorism.iyear.plot(kind = 'hist', bins = 30, figsize = (10,5), color = 'red', fontsize=15)
    plt.xlabel('Year', fontsize=15)
    plt.ylabel('Frequency', fontsize=15)
    plt.title(name)
    plt.show()
```

```python
for num in dic.keys():
    plot(num,dic[num])
```
